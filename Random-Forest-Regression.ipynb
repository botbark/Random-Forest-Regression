```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_squared_error
```

```python
df = pd.read_csv('RF-Regression-Data.csv')
df = df.head(24)
```

```python
df.shape
```

```python
x = df.x.values.reshape(-1, 1)
y = df.y.values.reshape(-1, 1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
RandomForestRegModel = RandomForestRegressor()
```

```python
RandomForestRegModel.fit(x_train,y_train)
```

```python
y_pred = RandomForestRegModel.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
from IPython.display import Image  
from sklearn import tree
import pydotplus
```

```python
i_tree = 0
for tree_in_forest in RandomForestRegModel.estimators_:
    if (i_tree ==3):        
        dotfile = tree.export_graphviz(tree_in_forest, out_file=None)
        graph = pydotplus.graph_from_dot_data(dotfile)        
    i_tree = i_tree + 1
Image(graph.create_png())
```
